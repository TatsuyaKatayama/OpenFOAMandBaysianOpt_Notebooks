# [最適化問題] damBreakの水を高く飛ばす

* 最大化問題 \
    $x_{opt}  = argmax\{f(x)\}$
* 目的関数 \
    0.4 [s]時における右端(x+)面での水の最高到達点($alpha.water \geq 0.5$)
* 設計変数
    $x = (h, w)$
* 制約条件
    $0.1*0.146 \leq h \leq 0.8*0.146$
    $1.8*0.146 \leq w \leq 2.1*0.146$

* Caseディレクトリは「[02.OpenFOAMの自動化](./02.OpenFOAMの自動化.ipynb)」で作成したdamBreakを用いること

![damBreak](./imgs/damBreak.png)

## 自動化の準備
自動化に必要な設定を実施する。今回は下記を実施。

```python
from PyFoam.RunDictionary.SolutionDirectory import SolutionDirectory

#Caseを指定する
damBreak = SolutionDirectory("./damBreak")
```

```python
from PyFoam.RunDictionary.ParsedBlockMeshDict import ParsedBlockMeshDict

#blockMeshDict.orgの文字列化したvertices
blockMeshDict0 = ParsedBlockMeshDict("{}.org".format(damBreak.blockMesh()))
vertices = blockMeshDict0.content["vertices"]
map_result = map(str, vertices)
str_result = ','.join(map_result)
```

## ヘルパー関数の作成

```python
# 実習2 OpenFOAMを実行する上で便利なヘルパー関数を事前に作っておく

from PyFoam.Execution.BasicRunner import BasicRunner

def ofExec(cmd, case_path):
    """
    OpenFOAMのコマンドを実行する

    Parameters
    ----------
    cmd : str
        OpenFOAMのコマンド
    case_path : str
        実行対象のcaseの場所
    """
    cmds = shlex.split("{} -case {}".format(cmd, case_path))
    runner = BasicRunner(cmds, silent=True)
    state = runner.start()
    return state

```

## 目的関数の作成


```python
#実習3 def auto_damBreakを完成させる

import numpy as np

def auto_damBreak(x):
    """
    blockMeshDictを変更後計算し、右端面の水の最高到達点を返す

    Parameters
    ----------
    x : numpy.array
        1行2列の配列
    """
    h = x[0,0]
    w = x[0,1]
    
    #まずはクリーニング（計算結果の削除）
    damBreak.clearResults()
    damBreak.clearPattern("PyFoam*")
    
    
    #blockMeshDictの準備
    str_edit = str_result.replace(
                                " 0.32876 ", " {} ".format(h)
                        ).replace(
                                "(2 ", "({} ".format(w)
                        )


    blockMeshDict0.content["vertices"] = str_edit.split(',')

    #blockMeshDictとして書き出す
    blockMeshDict0.writeFileAs(damBreak.blockMesh())
    
    
    #OpenFOAMコマンドの実行
    res_blockMesh = ofExec("blockMesh", damBreak.name)
    res_setFields = ofExec("setFields", damBreak.name)
    res_interFoam = ofExec("interFoam", damBreak.name)
    
    #計算結果の取得
    hwater = np.loadtxt(
        "{}/postProcessing/singleGraph/0.4/line_alpha.water.xy".format(
            damBreak.name))
    hwater[0,1] = 0.5
    res = hwater[hwater[:,1] >= 0.5,0].max()

    return res
```

下記コードを実行し、0.418563と表示されればOK.

```python
x = np.array([[0.3333,1.8]])
auto_damBreak(x)
```

## 設計変数の定義

```python
#実習4 boundsリスト内に設計変数を定義する

bounds =[{'name': 'h', 'type': 'continuous', 'domain': (0.1, 0.8)}, 
         {'name': 'w', 'type': 'continuous', 'domain': (1.8, 2.1)}]
```

## 最適化問題の定義


```python
import GPyOpt
from numpy.random import seed

seed(1) 

#実習5 最適化問題を定義し、初期実験5回、+25回の最適化計算を実行。

myProblem = GPyOpt.methods.BayesianOptimization(auto_damBreak,
    domain=bounds,
    initial_design_numdata=5,
    initial_design_type="latin",
    maximize=True)

max_iter = 25
myProblem.run_optimization(max_iter) 
```

## 最適化解の確認
最適化履歴をプロットし、解を表示する。

```python
print(myProblem.x_opt, myProblem.fx_opt)
```

```python
%matplotlib inline 
myProblem.plot_convergence() 
```
